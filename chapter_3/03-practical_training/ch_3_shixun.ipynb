## 实训1

```python
import numpy as np
import matplotlib.pyplot as plt
import os
os.listdir()
pop_data = np.load('populations.npz', allow_pickle=True)
print(pop_data.files) # 这个属性是为了查看有多少个file和他们的名字
print(pop_data['feature_names'], '\n') # 提取特征名称
print(type(pop_data['data'][:-2,:]))
pop_data_nonan = pop_data['data'][:-2,:]
#print(pop_data_nonan)
pop_data_sort = pop_data_nonan[pop_data_nonan[:,0].argsort(),:]
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.subplots_adjust(wspace=0.3, hspace=0.3)
x1 = plt.figure(figsize=(9, 9), dpi=80)
x1.add_subplot(2,1,1)
plt.plot(pop_data_sort[:,0],pop_data_sort[:,1])
plt.scatter(pop_data_sort[:,0],pop_data_sort[:,4])
plt.scatter(pop_data_sort[:,0],pop_data_sort[:,5])
plt.xlabel('年份')
plt.ylabel('年末总人口')
plt.xticks(range(0,20,2),pop_data_sort[range(0,20,2),0], rotation=45)
plt.title('1996～2015年全国人口图')
plt.legend(['总人口', '城镇人口', '乡村人口'])
x1.add_subplot(2,1,2)
#plt.plot(pop_data_sort[:,0],pop_data_sort[:,1])
plt.scatter(pop_data_sort[:,0],pop_data_sort[:,2])
plt.scatter(pop_data_sort[:,0],pop_data_sort[:,3])
plt.xlabel('年份')
plt.ylabel('年末总人口')
plt.xticks(range(0,20,2),pop_data_sort[range(0,20,2),0], rotation=45)
plt.title('1996～2015年全国人口图')
plt.legend(['总人口', '男性人口', '女性人口'])
plt.savefig('pop_plot1.png')
plt.show()
```

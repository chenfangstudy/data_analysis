## 掌握绘图基础语法与常用参数

```python
import numpy as np
import os
import matplotlib.pyplot as plt
#import matplotlib.mlab  as mlab
data = np.arange(0, 1, 0.01)
plt.title('lines')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0, 1))
plt.ylim((0, 1))
#np.arange(0, 1.1, 0.2)
(i for i in np.arange(0,1.1,0.2))
plt.xticks(np.arange(0,1.1,0.2))
plt.yticks(np.arange(0,1.1,0.2))
plt.plot(data, data**2)
plt.plot(data, data**4)
plt.legend(['y=x^2', 'y=x^4'])
plt.savefig('../tmp/pic1.png')
os.listdir('../tmp')
plt.show()
```

```python
rad = np.arange(0, np.pi*2,0.01)
# 第一副子图
p1 = plt.figure(figsize=(8, 6),dpi=80)
ax1 = p1.add_subplot(2,1,1)
plt.title('lines')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim((0,1))
plt.ylim((0,1))
plt.xticks(np.arange(0,1.1,0.2))
plt.yticks(np.arange(0,1.1,0.2))
plt.plot(rad, rad**2)
plt.plot(rad, rad**4)
plt.legend(['y=x^2','y=x^4'])
# di er fu zi tu
ax2 = p1.add_subplot(2,1,2)
plt.title('sin/cos')
plt.xlabel('rad')
plt.ylabel('value')
plt.xlim((0,np.pi*2))
plt.ylim((-1, 1))
plt.plot(rad, np.sin(rad))
plt.plot(rad, np.cos(rad))
plt.legend(['sin', 'cos'])
plt.savefig('../tmp/sincos1.png')
plt.show
os.listdir()
```

```python
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['lines.linestyle'] = '-.'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.marker'] = 'D'
x = np.linspace(0, 4*np.pi)
y = np.sin(x)
plt.plot(x,y,label='sin(x)')
plt.title('sin曲线')
plt.savefig('../tmp/sin20.png')
plt.show()
#plt.rcParams.__init__
```
